```python
import numpy as np
import pandas as pd
```

```python

df = pd.read_excel('/content/try3_Raw_Data_of_Chain_Pillar_try.xlsx')
df
```

```python
df=df.iloc[1:,1:]
```

```python
new_column_names = ['width of the pillar', 'Height of the pillar', 'Cover Depoth', 'Face length or Panel Width', 'Abutment angle',
                   '(Compressive Strength of the Coal) (MPa)', 'Modulus ratio of roof to coal', 'Modulus ratio of floor to coal', 'Caving Height (m)', 'Elastic modulus of Overburden (Gpa)','Density of the overburden(kg/m3)','Factor of Safety']  # Replace with your desired column names
df.columns = new_column_names
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

we can see that our data have no missing value

```python

import matplotlib.pyplot as plt

for column in df.columns[:-1]:
  plt.figure(figsize=(8, 6))
  plt.boxplot(df[column])
  plt.title(f"Box Plot of {column}")
  plt.ylabel(column)
  plt.show()
```

We can see  box plots of every data point, and there are outlier data in the abutment angle only. So we can drop the rows where outliers are present, or we can impute the maximum limit value for higher outliers and the minimum limit for lower outliers.

```python
df.head()
```

```python

Q1 = df['Abutment angle'].quantile(0.25)
Q3 = df['Abutment angle'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Abutment angle'] >= lower_bound) & (df['Abutment angle'] <= upper_bound)]
df_cleaned
```

```python
df_cleaned.info()
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

```python

column=df_cleaned.iloc[:,:]
for column in df.columns:
  plt.figure(figsize=(8, 6))
  plt.boxplot(df_cleaned[column])
  plt.title(f"Box Plot of {column}")
  plt.ylabel(column)
  plt.show()
```

We have cleaned the data there are no more outliers in input column

```python

plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['Factor of Safety'], bins=30, edgecolor='black')
plt.xlabel('Factor of Safety')
plt.ylabel('Frequency')
plt.title('Distribution of Factor of Safety')
plt.grid(True)
plt.show()
```

```python

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.distplot(df_cleaned['Factor of Safety'], bins=30, kde=True)
plt.xlabel('Factor of Safety')
plt.ylabel('Density')
plt.title('Distribution of Factor of Safety')
plt.grid(True)
plt.show()
```
